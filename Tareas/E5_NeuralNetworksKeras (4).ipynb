```python
import numpy as np
import keras
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

# 1ï¸âƒ£ Generar los datos
x_train, y_train = make_moons(n_samples=1000, noise=0.2, random_state=3)

# 2ï¸âƒ£ Crear el modelo de red neuronal
model = Sequential([
    Dense(10, input_shape=(2,), activation="relu"),  # Capa oculta con 10 neuronas
    Dense(10, activation="relu"),  # Otra capa oculta
    Dense(1, activation="sigmoid")  # Capa de salida (clasificaciÃ³n binaria)
])

# 3ï¸âƒ£ Compilar y entrenar
model.compile(optimizer=Adam(learning_rate=0.01), loss="binary_crossentropy", metrics=["accuracy"])
model.fit(x_train, y_train, epochs=50, verbose=1)

# 4ï¸âƒ£ FunciÃ³n para visualizar la frontera de decisiÃ³n
def plot_decision_region(model, X, pred_fun):
    min_x = np.min(X[:, 0]) - 0.1
    max_x = np.max(X[:, 0]) + 0.1
    min_y = np.min(X[:, 1]) - 0.1
    max_y = np.max(X[:, 1]) + 0.1
    x_vals = np.linspace(min_x, max_x, 200)
    y_vals = np.linspace(min_y, max_y, 200)
    XX, YY = np.meshgrid(x_vals, y_vals)
    ZZ = np.array([pred_fun(model, x1, x2) for x1, x2 in zip(XX.ravel(), YY.ravel())])
    ZZ = ZZ.reshape(XX.shape)
    plt.contourf(XX, YY, ZZ, cmap=plt.cm.coolwarm, alpha=0.7)
    plt.colorbar()

# 5ï¸âƒ£ FunciÃ³n para predecir
def pred_fun(model, x1, x2):
    return model.predict(np.array([[x1, x2]]), verbose=0)[0, 0]

# 6ï¸âƒ£ Graficar la frontera de decisiÃ³n y los datos
plt.figure(figsize=(10, 6))
plot_decision_region(model, x_train, pred_fun)
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, s=40, cmap=plt.cm.coolwarm, edgecolors='k')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Frontera de DecisiÃ³n - Red Neuronal en Keras")
plt.show()
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from torch.utils.data import DataLoader, TensorDataset

# 1ï¸âƒ£ Generar los datos
x_train, y_train = make_moons(n_samples=1000, noise=0.2, random_state=3)
x_train_torch = torch.tensor(x_train, dtype=torch.float32)
y_train_torch = torch.tensor(y_train, dtype=torch.float32).reshape(-1, 1)

# 2ï¸âƒ£ Crear el modelo de red neuronal en PyTorch
class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.hidden1 = nn.Linear(2, 10)
        self.hidden2 = nn.Linear(10, 10)
        self.output = nn.Linear(10, 1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.hidden1(x))
        x = self.relu(self.hidden2(x))
        x = self.sigmoid(self.output(x))
        return x

model = NeuralNet()

# 3ï¸âƒ£ Definir la funciÃ³n de pÃ©rdida y el optimizador
criterion = nn.BCELoss()  # Para clasificaciÃ³n binaria
optimizer = optim.Adam(model.parameters(), lr=0.01)

# 4ï¸âƒ£ Entrenar el modelo
num_epochs = 50
batch_size = 32
dataset = TensorDataset(x_train_torch, y_train_torch)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

for epoch in range(num_epochs):
    for batch_x, batch_y in dataloader:
        optimizer.zero_grad()
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()

# 5ï¸âƒ£ FunciÃ³n para visualizar la frontera de decisiÃ³n
def plot_decision_region(model, X, pred_fun):
    min_x, max_x = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    min_y, max_y = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    x_vals = np.linspace(min_x, max_x, 200)
    y_vals = np.linspace(min_y, max_y, 200)
    XX, YY = np.meshgrid(x_vals, y_vals)
    ZZ = np.array([pred_fun(model, x1, x2) for x1, x2 in zip(XX.ravel(), YY.ravel())])
    ZZ = ZZ.reshape(XX.shape)
    plt.contourf(XX, YY, ZZ, cmap=plt.cm.coolwarm, alpha=0.7)
    plt.colorbar()

# 6ï¸âƒ£ FunciÃ³n de predicciÃ³n
def pred_fun(model, x1, x2):
    x_tensor = torch.tensor([[x1, x2]], dtype=torch.float32)
    with torch.no_grad():
        return model(x_tensor).item()

# 7ï¸âƒ£ Graficar la frontera de decisiÃ³n y los datos
plt.figure(figsize=(10, 6))
plot_decision_region(model, x_train, pred_fun)
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, s=40, cmap=plt.cm.coolwarm, edgecolors='k')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Frontera de DecisiÃ³n - Red Neuronal en PyTorch")
plt.show()
```
